```python
import cv2
import time
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt
```

### Start time of execution

```python
start = time.time()
```

### 1. Read image

```python
easy_text_path = "images/easy_text.png"
easy_img = cv2.imread(easy_text_path)
```

```python
plt.imshow(easy_img)
plt.show()
```

### 2. Using the OCR

```python
text = pytesseract.image_to_string(easy_img)
print(text)
```

```python
data = pytesseract.image_to_data(easy_img, output_type=Output.DICT)
```

```python
n_boxes = len(data["text"])
```

### 3. Ploting the Bounding Boxes

```python
for i in range(n_boxes):
   if data["conf"][i] == -1:
       continue
   # Coordinates
   x, y = data["left"][i], data["top"][i]
   w, h = data["width"][i], data["height"][i]

   # Corners
   top_left = (x, y)
   bottom_right = (x + w, y + h)

   # Box params
   green = (0, 255, 0)
   thickness = 3  # pixels

   cv2.rectangle(img = easy_img, pt1 = top_left, pt2 = bottom_right, color = green, thickness = thickness)
```

```python
plt.imshow(easy_img)
plt.show()
```

### Time of execution

```python
end = time.time()
```

```python
delta = (end - start)

hours = int(delta/3600)
mins = int((delta - hours*3600)/60)
segs = int(delta - hours*3600 - mins*60)
print(f'Execute this notebook take us {hours} hours, {mins} minutes and {segs} seconds.')
```
