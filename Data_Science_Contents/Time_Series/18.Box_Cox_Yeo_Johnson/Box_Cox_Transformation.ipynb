```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
```

```python
def display_img(name, a, b):
    plt.figure(figsize = (a, b))
    image = mpimg.imread(name)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

# 1. Box Cox

```python
display_img('boxcox.png', 7, 7)
```

The goal of Box-Cox transformation is change the distribution of a random variable to Normal. To do this we maximize the likelihood asuming Normality.

### i. Exponential

```python
original_data = np.random.exponential(size = 1000)
```

```python
fitted_data, fitted_lambda = stats.boxcox(original_data)
```

```python
fitted_lambda
```

```python
plt.figure(figsize = (10, 7))

pd.Series(original_data).hist(bins = 30, color = 'blue', density = True)

plt.xlabel('Original Data', fontsize = 17)

plt.show()
```

```python
plt.figure(figsize = (10, 7))

pd.Series(fitted_data).hist(bins = 15, color = 'orange', density = True)

plt.xlabel('Transformed Data', fontsize = 17)

plt.show()
```

### ii. Exp(Normal)

```python
original_data = np.exp(np.random.normal(size = 1000))
```

```python
fitted_data, fitted_lambda = stats.boxcox(original_data)
```

```python
fitted_lambda
```

```python
plt.figure(figsize = (10, 7))

pd.Series(original_data).hist(bins = 30, color = 'blue', density = True)

plt.xlabel('Original Data', fontsize = 17)

plt.show()
```

```python
plt.figure(figsize = (10, 7))

pd.Series(fitted_data).hist(bins = 30, color = 'orange', density = True)

plt.xlabel('Transformed Data', fontsize = 17)

plt.show()
```

# 2. Yeo Johnson

```python
display_img('yeojohnson.png', 7, 7)
```

The Yeo Johnson transformation is used when the distribution have negative values.

### i. Normal

```python
original_data = np.random.normal(size = 1000)
```

```python
fitted_data, fitted_lambda = stats.yeojohnson(original_data)
```

```python
fitted_lambda
```

```python
plt.figure(figsize = (10, 7))

pd.Series(original_data).hist(bins = 30, color = 'blue', density = True)

plt.xlabel('Original Data', fontsize = 17)

plt.show()
```

```python
plt.figure(figsize = (10, 7))

pd.Series(fitted_data).hist(bins = 15, color = 'orange', density = True)

plt.xlabel('Transformed Data', fontsize = 17)

plt.show()
```
