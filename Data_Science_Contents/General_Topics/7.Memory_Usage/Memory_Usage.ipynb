# Dataframes Issues

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'column_1': [1, 2], 'column_2': [3, 4]})
```

```python
df
```

```python
df_2 = df
```

```python
df_2.iloc[1, 1] = 10
```

```python
df_2
```

```python
df
```

**Comments:** You can see when you asssign "df_2 = df" you only create a new variable ("df_2") with the same memory access than "df". For this reason when you change elements in "df_2" then "df" changes too. The true way to copy a dataframe is:

```python
df_3 = df.copy()
```

```python
df
```

```python
df_3
```

```python
df_3.iloc[0, 0] = 11
```

```python
df_3
```

```python
df
```
