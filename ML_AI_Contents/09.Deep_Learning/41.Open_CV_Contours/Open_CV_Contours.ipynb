```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import cv2
import time
import matplotlib.pyplot as plt
from skimage import img_as_ubyte
```

### Start time of execution

```python
start = time.time()
```

# Execution

### 1. Open image

```python
img = cv2.imread("img.png")
```

```python
plt.imshow(img)
plt.show()
```

### 2. Convert to gray

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap = 'gray')
plt.show()
```

### 3. Binarize the image

```python
img_bin = (gray >= 230)*255
img_bin = img_as_ubyte(img_bin)
```

```python
plt.imshow(img_bin, cmap = 'gray')
plt.show()
```

### 4. Extract the contour

```python
contours, heirarchy = cv2.findContours(img_bin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

### 5. Plot the contour

```python
_ = cv2.drawContours(img, contours, -1, (0,255,0), 20)
```

```python
plt.imshow(img)
plt.show()
```

### Time of execution

```python
end = time.time()
```

```python
delta = (end - start)

hours = int(delta/3600)
mins = int((delta - hours*3600)/60)
segs = int(delta - hours*3600 - mins*60)
print(f'Execute this notebook take us {hours} hours, {mins} minutes and {segs} seconds.')
```
