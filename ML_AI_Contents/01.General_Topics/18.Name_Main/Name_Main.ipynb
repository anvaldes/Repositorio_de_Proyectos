When we import modules on occasions we execute different functions in the modules, but we only want import the functions and not the executions of its functions on the module its come from. This is the purpose why we use the line: '__name__=='__main__'.

# Example 1

We build a file called "file_1.py" and this is the code:

```python
import file_1
```

```python
file_1.__name__
```

**Comments:** As you can see, when we import the module also its execute all the functions we called on the module file.

# Example 2

We build a file called "file_2.py" and this is the code:

```python
import file_2
```

**Comments:** As you can see, when we import the module this time doesn't execute all the functions we called on the module file. This happens because when we call the module the variable "__name__" is not longer "__main__" in this occasion the variable "__name__" of the file "file_2.py" is "file_2".

```python
file_2.__name__
```
