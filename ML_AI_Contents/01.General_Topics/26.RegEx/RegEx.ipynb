A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

RegEx can be used to check if a string contains the specified search pattern.

```python
from matplotlib import image as mpimg
import matplotlib.pyplot as plt
```

```python
def display_img(name, a, b):
    plt.figure(figsize = (a, b))
    image = mpimg.imread(name)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
import re
```

# 2. Example

Search the string to see if it starts with "The" and ends with "Spain":

```python
txt = "The rain in Spain"
```

```python
x = re.search("^The.*Spain$", txt)
```

```python
if x:
    print('True')
else:
    print('False')
```

# 3. RegEx Functions

**findall**: Return a list containing all matches.

**search:** Regurn a "Match object" if there is a match anywhere in the string.

**split:** Returns a list where the string has been split at each match.

**sub:** Replaces one or many matches with a string.

### i. findall

```python
txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```python
txt = "The rain in Spain"
x = re.findall("oi", txt)
print(x)
```

### ii. search

```python
txt = "The rain in Spain"
x = re.search("\s", txt)
print(x.start())
```

```python
txt = "The rain in Spain rain"
x = re.search("rain", txt)
print(x.start())
print(x.end())
print(txt[x.start(): x.end()])
```

```python
txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x)
```

### iii. split

```python
txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```

```python
txt = "The rain in Spain"
x = re.split("\s", txt, 1) # Only the first occurrence
print(x)
```

### iv. sub

```python
txt = "The rain in Spain"
x = re.sub("\s", "9", txt)
print(x)
```

```python
txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2) # The first two occurrences
print(x)
```

```python
txt = "The rain in Spain"
x = re.sub("b", "9", txt, 2)
print(x)
```
