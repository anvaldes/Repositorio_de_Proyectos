# Explanation

$ p(x) = \sum_{c} \pi_c \cdot N(x | \mu_c, \sigma_c) $

$ r_{i, c} = \cfrac{\pi_c \cdot N(x | \mu_c, \sigma_c)}{\sum_{c'} \pi_c' \cdot N(x | \mu_c', \sigma_c')} $

For each point $(i)$ choose the cluster $(c)$ with the highest $ r_{i, c} $.

### M step

For each $ c $ (cluster):

$ \ $

$ m_c = \sum_{i} r_{i, c} $

$ \pi_c = \cfrac{m_c}{m} $

$ \mu_c = \cfrac{1}{m_c} \sum_{i} r_{i, c} \cdot x^{(i)} $

$ \Sigma_c = \cfrac{1}{m_c} \sum_{i} r_{i, c} \cdot (x^{(i)} - \mu_c)^T (x^{(i)} - \mu_c) $

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
df = pd.DataFrame(X)
```

```python
plt.scatter(df[0], df[1])
plt.show()
```

```python
gmm = GaussianMixture(n_components = 3)
```

```python
gmm = gmm.fit(df)
```

```python
labels = gmm.predict(df)
df['labels']= labels
```

```python
df_0 = df[df['labels']== 0]
df_1 = df[df['labels']== 1]
df_2 = df[df['labels']== 2]
```

```python
plt.scatter(df_0[0], df_0[1], c ='red')
plt.scatter(df_1[0], df_1[1], c ='blue')
plt.scatter(df_2[0], df_2[1], c ='green')
plt.show()
```
