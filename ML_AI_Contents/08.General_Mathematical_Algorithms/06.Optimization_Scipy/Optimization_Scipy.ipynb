```python
import time
import math
import scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib.pyplot import figure
from matplotlib import image as mpimg
```

# Execution

### 1. Quadratic

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.1, 3.5, 7.2, 13.1, 21.3])
```

```python
def func(x, a, b, c):
    return a * x**2 + b * x + c
```

```python
popt, pcov = curve_fit(func, x, y)
```

```python
a, b, c = popt
```

```python
plt.plot(x, y, 'ro', label='Datos')
plt.plot(x, func(x, a, b, c), label='Ajuste')
plt.legend()
plt.show()
```

### 2. Exponential

```python
def func(x, a, b, c):
    return a*np.exp(-b*x) + c
```

```python
xdata = np.linspace(0, 4, 50)
ydata = func(xdata, 2.5, 1.3, 0.5) + 0.2*np.random.normal(size=len(xdata))
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
plt.plot(xdata, ydata, 'o', label='Data')
plt.plot(xdata, func(xdata, *popt), '-', label='Fitted Curve')
plt.legend()
plt.show()
```

### 3. Logarithm

```python
def func(x, a, c):
    return a*np.log(x) + c
```

```python
xdata = np.linspace(1, 4, 50)
ydata = func(xdata, 2.7, 1.7) + 0.2*np.random.normal(size=len(xdata))
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
plt.plot(xdata, ydata, 'o', label='Data')
plt.plot(xdata, func(xdata, *popt), '-', label='Fitted Curve')
plt.legend()
plt.show()
```
